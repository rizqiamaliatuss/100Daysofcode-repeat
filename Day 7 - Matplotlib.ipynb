```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

font1 = {'family':'serif','color':'blue','size':20} #created font
font2 = {'family':'serif','color':'darkred','size':15}

plt.title('the data', fontdict = font1, loc = 'left')
plt.xlabel('xlabel', fontdict = font2) #labeling
plt.ylabel('ylabel', fontdict = font2)


plt.grid() # add grid
plt.show()
```

# SCATTER PLOT

```python
# plot 1:
x1 = np.array([0, 1, 2, 3,5,6,8,3,14,16,17])
y1 = np.array([3, 8, 1, 10,5,8,9,10,18,30,11])

plt.scatter(x1,y1)

plt.show()
```

```python
# plot 1:
x1 = np.array([0, 1, 2, 3,5,6,8,3,14,16,17])
y1 = np.array([3, 8, 1, 10,5,8,9,10,18,30,11])
plt.scatter(x1,y1)

x1 = np.array([6,7,3,6,7,5,1,3,6,9,5,17])
y1 = np.array([3,4,5,6,2,3,4,6,7,8,3,4])
plt.scatter(x1,y1, color = 'hotpink')

plt.show()
```

```python
x1 = np.array([6,7,3,6,7,5,1,3])
y1 = np.array([10,20,30,40,50,60,70,80])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55])
plt.scatter(x1,y1,c=colors, cmap = 'winter')
plt.colorbar()
plt.show()
```
